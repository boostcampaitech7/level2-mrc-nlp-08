```python
import json
filename='ensemble'
with open(f'{filename}.json', 'r', encoding='utf-8') as f:
        data = json.load(f)
```

```python
data
```

```python
def split_and_remove_duplicates(dictionary):
    for key, value in dictionary.items():
        # Split the value by space
        words = value.split()
        # Remove duplicates while keeping the order
        unique_words = list(dict.fromkeys(words))
        # Join the unique words back into a string
        dictionary[key] = ' '.join(unique_words)
    return dictionary
```

```python
updated_data = split_and_remove_duplicates(data)
```

```python
updated_data
```

```python
def save_to_json(data, output_file):
    with open(output_file, 'w', encoding='utf-8') as f:
        json.dump(data, f, ensure_ascii=False, indent=4)
```

```python
save_to_json(updated_data, f'ensemble_source/{filename}_dup_remove_best_texts.json')
```
