```python
import pandas as pd
```

```python
ensemble_list = [
    "nbest_predictions_버트_BM25_best_texts",
    "nbest_predictions_버트_CNN_best_texts",
    "nbest_predictions_버트_dpr_best_texts",
    
    "sudong_clean_bm25_nbest_predictions_best_texts",
    "sudong_clean_final_bm25_nbest_predictions_best_texts",
    "sudong_sudong_clean_final_bm25_dpr_nbest_predictions_best_texts",
    
    "nbest_predictions_티나_bm25_best_texts",
    "nbest_predictions_티나_dpr_best_texts",
    "nbest_predictions_티나_CNN_best_texts",
    
    "sudong_bm25_best_texts",
    "sudong_origin_best_texts",
    "sudong_dpr_best_texts",
    # 홍조 
    # 수동 CNN만
    # "nbest_predictions_수동_top40_bm25_best_texts",
    "nbest_predictions_수동_top40_dpr+bm25_best_texts",
    
    # "nbest_predictions_hongjo_best_texts",
    "nbest_predictions_hongjo_bm25_best_texts",
    # "nbest_predictions_hongjo_dpr_best_texts",
    
    # "noah_origin_best_texts",
    # "noah_2tina_best_texts",
    # "noah_bm25_best_texts",
    
    # "nbest_predictions_line_best_texts",
    "nbest_predictions_line_bm25_best_texts",
    # "nbest_predictions_line_dpr_best_texts"
]
```

```python
data_list = []
# 파일 읽어서 리스트에 저장
for i, case in enumerate(ensemble_list):
    # 파일 경로에 확장자를 추가하고 데이터를 읽음
    file_path = f"ensemble_source/{case}.json"
    data = pd.read_json(file_path).transpose()
    data_list.append(data)
    
# print(data_list[0].head())
    
```

```python
for i in range(len(data_list)):
    print(len(data_list[i]))
```

```python
anslist = []
for i in range(len(data_list[0])):
    anslist.append([(data_list[j].iloc[i]['best_text'], data_list[j].iloc[i]['probability']) for j in range(len(data_list))])
```

```python
columns = ensemble_list
ans_df = pd.DataFrame(anslist, index=data_list[0].index, columns=columns)
```

```python
ans_df
```

```python
from collections import Counter

def select_text(row):
    # 모든 텍스트와 확률을 각 모델별로 추출
    texts = [row[model][0] for model in ensemble_list]
    probs = [row[model][1] for model in ensemble_list]
    
    # 텍스트 빈도 계산
    text_counter = Counter(texts)
    most_common_text, _ = text_counter.most_common(1)[0]
    v_count = list(text_counter.values())
    
    if len(v_count) == 1:
        return most_common_text
    elif v_count[0] != v_count[1]:
        return most_common_text
    elif v_count[0] == v_count[1]:
        max_prob_index = probs.index(max(probs))
        return texts[max_prob_index]
    else:  # 다수결이 없을 경우, 확률이 가장 큰 값을 선택
        max_prob_index = probs.index(max(probs))
        return texts[max_prob_index]
```

```python
result_df = pd.DataFrame(ans_df.apply(select_text, axis=1), index=data_list[0].index, columns=['result'])
```
